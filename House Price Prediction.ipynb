```python
def convert_amount(value):
    if pd.isna(value):
        return np.nan
    
    value = value.strip()
    
    if 'Cr' in value:
        return float(value.replace('Cr', '').strip()) * 10000000
    elif 'Lac' in value:
        return float(value.replace('Lac', '').strip()) * 100000
    else:
        return np.nan

df['Amount(in rupees)'] = df['Amount(in rupees)'].apply(convert_amount)
```

```python
df['Amount(in rupees)'].head()
```

```python
df['Carpet Area'] = (
    df['Carpet Area']
    .str.replace('sqft', '', regex=False)
    .str.strip()
)

df['Carpet Area'] = pd.to_numeric(df['Carpet Area'], errors='coerce')
```

```python
def extract_floor(value):
    if pd.isna(value):
        return np.nan, np.nan
    
    value = value.lower().strip()

    # Handle basement floors
    if 'basement' in value:
        return -1, np.nan   # basement as -1

    # Handle ground floor
    if value.startswith('ground'):
        if 'out of' in value:
            total = value.split('out of')[1].strip()
            return 0, int(total)
        return 0, np.nan

    # Handle normal "X out of Y"
    if 'out of' in value:
        current, total = value.split('out of')
        if current.strip().isdigit() and total.strip().isdigit():
            return int(current.strip()), int(total.strip())

    # Anything else → unknown
    return np.nan, np.nan
```

```python
df[['current_floor', 'total_floors']] = df['Floor'].apply(
    lambda x: pd.Series(extract_floor(x))
)
```
